`Lists` 
Lists unlike arrays are not indexed. They are used to store collections of items that are not ordered. They are also known as sets or collections. Lists are mutable, meaning they can be changed after they are created. They are also dynamic, meaning they can grow or shrink as items are added or removed.   
Lists are defined by square brackets `[]` and elements are separated by commas. Here is an example of a list: `[1, 2, 3, 4, 5]`.
