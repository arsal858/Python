`Tuples`

```python
# empty tuple
t1=tuple()
print(t1)  # Output: ()
# tuple on single element
t2=(2)
print(type(t2))
# Output: <class 'int'>
# tuple on multiple elements
t3=(1,2,3)
```

```python
# hetrogenious tuple
t4=(1,2,3,'a','b','c',True,None)
print(t4)
a,b,*others=(1,2,3,4)
print(a,b)
print(others)
```

```python
# tuple unpacking
t5=(1,2,3,'a','b','c',True,None)
a,b,c,d,e,f,g,h=t5
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
```

```python
# indexing and slicing of tuples
t4[0]
t4[0:3]
# inversing a tuple
t4[::-1]
```

```python
# editing a tuple
# tuples like strings are immuteable
t4[1]=10
```

```python
# deleting a tuple : a portion cannot be deleted
del t3 # entire tuple can be deleted
```

```python
# operations and functions in tuples
print(t4+t5)
```

`Sets`

```python
# curly braces
s={} # it is a dictionary not a set
s1={1,2,3}
print(s1)
```

```python
# 2d sets are not allowed
s2={1,2,3,4,True,None}
print(s2)
```

```python
# we can add tuples inside it
s3={1,3,4,(1,6)}
print(s3)
```

```python
s5=set([1,5,9,9])
```

```python
# accessing the set ---- does not work
# indexing is not allowed
# adding an element
s5.add(6)
s5.update([1,4,5])
```

```python
s5
```

```python
# deleting and deleting an item of a set
del a
```

```python
s5.discard(5) # deleting an item from the set
s5.remove(9) # deleting an item from the set
print(s5)
```

```python
s5.pop() # randomly delete an item
```

```python
s5.clear() # clear an item
```

```python
# set operations
s1={1,4,5}
s2={1,6,7}
# union
s1|s2
# intersection
s1&s2
s1-s2 # the items of s1 that are not present in s2
# symmetric difference
s1^s2 # the items that are in s1 or s2 but not in both
1 in s1
```

```python
# set functions
len(s1) # returns length
sum(s1) # returns sum of all elements
min(s1) # returns smallest element
max(s1) # returns largest element
sorted(s1) # returns sorted list of elements
s1.union(s2) # returns union of two sets
s1.update(s2) 
```

```python
# subset and superset
s1={1,3,5,7}
s2={11,14,16,1}
s1.superset(s2)
```

```python
# frozenset
s=frozenset({1,3,6})

s1=frozenset({1,3,5,frozenset({1,6})})

s1
```

**`Dictionaries`**

```python
# keys cannot be duplicate
# keys must be immutable
# values can be mutable
# keys are used for fast lookups, values are used for fast insertions and deletions
d=dict() # empty dictionary
```

```python
# 1d dictionary
d = {'a': 1, 'b': 2, 'c': 3}
print(d)

# 2d dictionary
d = {'a': {'x': 1, 'y': 2}, 'b': {'x': 3, 'y': 4}}
print(d)
```

```python
# we can pass list of tuples
# we can pass list of dictionaries
# we can pass list of lists
# we can pass list of strings
# we can pass list of integers
# we can pass list of floats
# we can pass list of booleans
# we can pass list of None
# we can pass list of custom objects
# we can pass list of custom objects with attributes
# we can pass list of custom objects with methods
a=dict([('name','arslan'),('age',20)])
print(a)
```

```python
a['name'] # fetching dictionary item
```

```python
a.get('name') # get the name of the key
```

```python
a['a']=40 # adding an item in dictioanry
```

```python
a.pop('a') # deleting an item
```

```python
a.popitem() # deleting an item(last item)
```
