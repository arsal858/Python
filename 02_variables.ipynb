`Operators in Python`
- Arthimatic Operators
- Logical Operators
- Relational Operators
- Biswise Operators
- Assignment Operators
- Membership Operators

```python
# arthimatic operators
print(5+4) # additition
print(5-4) # substraction
print(5*4) # multiplication
print(5/2) # division
print(5//2) # integer
print(5%2) # modulus
print(5**2) # power 
```

```python
# Relational operators for comparison
print(5>4) # greater than
print(5<4) # less than operator
print(5<=4) # less than equal to operator
print(5<=8) # greater than equal to operator
print(5==9) # equal equal to operator
print(4!=7) # not equal to operator
```

```python
# logical operators 
print(1 and 0)
print(1 or 0)
print(not 0)
```

```python
# bitwise operators not useful in datascience but image processing may be used
print(2 & 1) 
print(2|1)
print(2^3) # xror operator
print(~3) # bitwise not
print(4 >>2)  # bitwise left
print(5<<4) # bitwise right operators

```

```python
# Assignment Operators
a=3 
a+=3 # a=a+3
print(a)
```

```python
# Membership Operator
print(3 in [1,2,3])
print(3 not in [1,2,3])
print('I' in 'Islamabad')
```

```python
# break and continue statement.they are used for loop control.
# break statement is used to terminate the loop. data base query is an example of break statement.
for i in range(1,10):
    if i==5:
        break
    print(i)
    
for i in range(1,10):
    if i==5:
        continue
    print(i)
```

```python
lower=int(input('Please enter in the number'))
upper=int(input('Please enter in the number'))
for i in range(lower,upper+1):
    if i%2!=0:
        continue
    print(i)
```
